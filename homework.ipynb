### Problem 13

Set `p13` to the average $L_1$ loss using the average of 1989 and 1990 vineyard yields per row to predict 1991 yields per row.

```python
# YOUR CHANGES HERE
# L1 is the sum of absolute differences between pred and target
pred = (vineyard["lugs_1989"] + vineyard["lugs_1990"]).divide(2)
l1 = (vineyard["target"] - pred).abs()
p13 = l1.sum() / len(l1)
```

### Problem 15

Given the following data, set `p15` to the weighted variance of 

| Color | Shape | Score | Probability |
|---|---|---|---:|
| red | square | 3 | 0.250 |
| blue | circle | 4 | 0.125 |
| purple | line | 2 | 0.125 |
| purple | diamond | 5 | 0.25 |
| blue | triangle | 3 | 0.25 |

```python
# YOUR CHANGES HERE
# weighted variance:
# find weighted mean, compute squared difference from each point to mean, mult each squared difference by its weight, sum weighted squared differences
# weighted mean:
# compute sum of each val mult by its weight, divided by the sum of all the weights
p15_df = pd.DataFrame(
    {
        'Color': ['red', 'blue', 'purple', 'purple', 'blue'],
        'Shape': ['square', 'circle', 'line', 'diamond', 'triangle'],
        'Score': [3, 4, 2, 5, 3],
        'Probability': [.250, .125, .125, .25, .25],
    }
)
p15_df['Score_Prob'] = p15_df['Score'] * p15_df['Probability']
weights_sum = sum(p15_df['Probability']) # 1
weighted_mean = sum(p15_df['Score_Prob']) / weights_sum # 3.5
p15 = (((p15_df['Score'] - weighted_mean) ** 2) * p15_df['Probability']).sum()
```

```python
p15
```
